# Rapport Final : Prédiction de l'Attrition des Employés Utilisant l'`Arbre de Décision` et la `Forêt Aléatoire`

## 1. Résumé

Ce projet se concentre sur la prédiction de l'`attrition des employés` en utilisant des modèles de classification : l'`Arbre de Décision` et la `Forêt Aléatoire`. Les modèles ont été appliqués à un ensemble de données d'`attributs des employés` pour identifier ceux susceptibles de quitter l'organisation. L'étude vise à aider les organisations à mettre en œuvre des stratégies de rétention et à minimiser la perte de talents.

## 2. Introduction

### 2.1 Contexte
L'`attrition des employés` peut avoir un impact significatif sur la performance organisationnelle. La prédiction de l'`attrition` permet des mesures de rétention proactives, assurant la stabilité et la compétence de la main-d'œuvre. Les algorithmes d'`Arbre de Décision` et de `Forêt Aléatoire` sont largement utilisés pour les tâches de classification en raison de leur simplicité et de leur efficacité.

### 2.2 Objectifs
* Construire et évaluer des modèles d'`Arbre de Décision` et de `Forêt Aléatoire` pour la prédiction de l'`attrition des employés`
* Comparer la performance des deux modèles
* Interpréter l'importance des caractéristiques pour en tirer des insights actionnables

## 3. Méthodologie

### 3.1 Jeu de Données
* **Source** : `données projet2`
* **Caractéristiques** : Divers attributs des employés (`Age`, `EnvironmentSatisfaction`, `HourlyRate`, `Gender` etc.)
* **Variable Cible** : Statut d'`attrition` (`Oui/Non`)

### 3.2 Prétraitement
* **Analyse Exploratoire des Données** : Visualisation des distributions et relations entre caractéristiques
* **Nettoyage des Données** : Traitement des valeurs manquantes
* **Encodage** : Application des techniques d'encodage pour les variables catégorielles
* **Feature Creation** : nous effectuons une création de caractéristiques (`Feature Engineering`) en créant deux types d’interactions : **`Termes Quadratiques`** et **`Termes d’Interaction`**
* **Data Transformation** : `Standardization`
* **Division Train-Test** : Répartition des données en ensembles d'entraînement (`75%`) et de test (`25%`)

### 3.3 Modèles Implémentés

#### `Arbre de Décision`
* **Critère** : `Indice de Gini` / `Entropie`

    `'criterion'`,
    `'class_weight'`
#### `Forêt Aléatoire`

## 4. Métriques d'Évaluation
La performance a été mesurée en utilisant :
* **Précision (`Accuracy`)**
* **`Recall`**
* **Score `F1`**

## 5. Résultats et Analyse

### 5.1 Performance de l'`Arbre de Décision`

### 5.2 Performance de la `Forêt Aléatoire`

|----------|---------|

| Précision | `0.85` |
| `Recall` | `0.86` |
| Score `F1` | `0.85` |



### 5.3 Importance des Caractéristiques
**Les caractéristiques les plus influentes étaient** : `[Age, MonthlyIncome, HourlyRate]`.

## 6. Discussion

### 6.1 Comparaison des Modèles
La `Forêt Aléatoire` a constamment surpassé l'`Arbre de Décision` sur toutes les métriques grâce à sa nature d'ensemble, qui réduit le surapprentissage. Cependant, l'`Arbre de Décision` a fourni une interprétabilité plus facile.

### 6.2 Implications Pratiques
Les organisations peuvent utiliser le modèle de `Forêt Aléatoire` pour des prédictions de haute précision, tandis que le modèle d'`Arbre de Décision` offre des insights rapides et interprétables pour les décisions immédiates.

## 7. Conclusion
* **Meilleur Modèle** : La `Forêt Aléatoire` a obtenu une performance supérieure
* **Recommandations** : Utiliser le modèle de `Forêt Aléatoire` pour des prédictions précises et prioriser les caractéristiques importantes identifiées pour les stratégies de rétention
